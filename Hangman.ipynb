```python
from IPython.display import clear_output
import random

class Game():
    game_words = {
        'sport' : {
            1 : 'aerobics',
            2 : 'decathlon',
            3 : 'frisbee',
            4 : 'pingpong',
            5 : 'canoe'
        },
        'animal' : {
            1 : 'amoeba',
            2 : 'dragonfly',
            3 : 'flounder',
            4 : 'hedgehog',
            5 : 'meerkat'
        },
        'food' : {
            1 : 'casserole',
            2 : 'gelatin',
            3 : 'rhubarb',
            4 : 'prune',
            5 : 'meringue'
        }
    }
    
    def __init__(self):
        self.word = []
        self.flag = False
        
    def getWord(self):
        print('Categories')
        print('============')
        print('1. Sport')
        print('2. Animal')
        print('3. Food')
        ans = input('What category would you like to play? (enter number) ')
        if ans == '1':
            self.word = self.game_words['sport'][random.randint(1,5)]
        elif ans == '2':
            self.word = self.game_words['animal'][random.randint(1,5)]
        elif ans == '3':
            self.word = self.game_words['food'][random.randint(1,5)]
    
    def showPuzzle(self, player):
        if player.letter in self.word and len(player.letter) > 0:
            player.corrected.append(player.letter)
            self.flag = True

        for i in range(len(self.word)):
            if self.word[i] in player.corrected and len(player.letter) > 0:
                print('{} '.format(self.word[i]), end='')
            else:
                print('_ ', end='')
        print('')
        
        if not self.flag and player.lives > 0 and len(player.letter) > 0:
            player.lives -= 1
            print('\nYou have {} lives left!'.format(player.lives))   
    
        self.flag = False
        
    def gameOver(self, player):
        if player.lives == 0:
            return True
    
    def playerWin(self, player):
        for i in range(len(self.word)):
            if self.word[i] not in player.corrected:
                break
        else:
            return True
    
    def __str__(self):
        print(self.word)
        
class Player():
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    
    def __init__(self):
        self.guess_list = []
        self.letter = ''
        self.lives = 6
        self.corrected = []
    
    def guessLetter(self):
        ans = input('Select a letter (enter \'quit\' to end): ').lower()
        if ans == 'quit':
            return True
        elif ans in self.alphabet and ans not in self.guess_list:
            self.guess_list.append(ans)
            self.letter = ans
        elif ans not in self.alphabet:
            print('It is not a letter, please select again.')
            return False
        elif ans in self.guess_list:
            print('\'{}\' was already selected, please select again.'.format(ans))
            return False
    
    def __str__(self):
        print(self.guess_list)
        
while True:
    clear_output()
    
    game = Game()
    game.getWord()
    player = Player()
    
    clear_output()
    game.showPuzzle(player)
    
    while True: 
        decision = player.guessLetter()

        if decision:
            print('Thank you for playing, come back again!')
            break
        elif decision == False:
            player.letter = ''
            continue
        
        clear_output()
        
        print('Letters selected: ', end='')
        print(player.guess_list)
        game.showPuzzle(player)
        
        game_over = game.gameOver(player)
        player_win = game.playerWin(player)

        if game_over:
            print('\nSorry, you have been hanged! The answer is \'' + game.word + '\'.')
            print('Keep trying!!')
            break
        elif player_win:
            print('\nYou win!')
            break
           
    ans = input('Would you like to play again (y/n)? ')
    
    if ans.lower() == 'n':
        break
```
